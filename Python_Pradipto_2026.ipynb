#Recursion :
When a function calls itself

```python
# calculate the factorial of a number

# 5! = 5x4x3x2x1 = 120

# 5! = 5 x (5-1)!
#    = 4 x 3!
# n! = n x (n-1)!

# recursive case, base case(terminating case)
def pradipto(n):
  #base case
  if n==1:
    return 1
  #rec case
  return n * pradipto(n-1)

n = int(input("Enter the number:  "))
res = pradipto(n)
print("The result is :  ", res)
```

#OOP basics

```python
# attributes
class Student:
  def __init__(self, name, roll, marks):
    print("Init method is automatically called....")
    self.name = name
    self.roll = roll
    self.marks = marks
  def study(self, sub1, sub2):
    print(f"{self.name} is studying...{sub1} and {sub2}")
  def play(self, a, b, c):
    print(f"{self.name} is playing with numbers and the sum is {a+b+c}")

```

```python
# object creation

# init method is automatically called whenever we create a new object
# to create objects we use constructor - it is the class name along with attributes

s1 = Student("Kirshna", 101, 99.5) #constructor
s2 = Student("Anjani", 201, 98.3)
s3 = Student("Aradhna", 301, 99.7)
```

```python
# how to access the attributes.   '  obj_name.attr  '

print(s1.name)
print(s2.marks)
print(s3.roll)
```

#Polymorphism and Inheritance

```python
class Mobile:
  def __init__(self, owner, price, year):
    print("init method of MOBILE class called ....")
    self.owner = owner
    self.price = price
    self.year = year

  def call(self):
    print("Calling a friend.... using Mobile")
```

```python
#child class of Mobile
class iPhone(Mobile):
  def __init__(self, model, price):
    print("init method of iPhone class called....")
    self.model = model
    self.price = price
  def call(self):
    print("Calling someone using IPHONE....")

  # pass #to leave the indented block empty we use the pass keyword
```

#Recursion

Factorial of a number

```python
def fact(number):
  if number == 1 or number == 0:
    return 1
  else:
    return number * fact(number-1)

n = int(input("Enter the number:  "))

res = fact(n)
print(f"The factorial of {n} is {res}")
```

Fibonacci Series

```python
def fibo(number):
  if number == 1 or number == 0:
    return number
  else:
    return fibo(number-1) + fibo(number-2)

n = int(input("Enter the number of terms:  "))


for i in range(n):
  res = fibo(i)
  print(res, end=" ")
# print(f"The {n}th term in fibonacci series is {res}")
```

Armstrong number using Logic

```python
n = int(input("Enter the number to check whether it is Armstrong Number:   "))

orig = n

#calculate number of digits
d=0
while n>0:
  d+=1 #d = d + 1
  n = n//10

print(f"The number of digits in {orig} is {d}")

n = orig

res = 0
while n>0:
  r = n%10
  res += (r**d)
  n=n//10

print(f"res = {res}")
if res == orig:
  print(f"The number {orig} is Armstrong Number")
else:
  print(f"The number {orig} is not Armstrong Number")
```

armstrong number in Python

```python
n = input("Enter the number to check whether it is Armstrong Number:   ") #string
d = len(n)

res = 0
for i in n:
  res += int(i)**d

print("Yes, Armstrong") if res==int(n) else print("Not Armstrong")
```
